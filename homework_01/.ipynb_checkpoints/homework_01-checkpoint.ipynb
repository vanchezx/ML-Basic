## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
inp_string = input("Введите число: ")
if len(inp_string) == 5:
    new_string = inp_string[0]+inp_string[3:0:-1]+inp_string[4]
    print(new_string)
else:
    print("Введено не пятизначное число!")
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
inp_days = input("Введите кол-во дней: ")
if inp_days.isdigit():
    days = int(inp_days)
    weekend = int(days) // 7 * 2
    if days % 7 == 6:
        weekend = weekend + 1
    print(weekend)
else:
    print("Введены не числовые данные!")
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
ch_len = int(input('Введите длину плитки: '))
ch_wid = int(input('Введите ширину плитки: '))
ch_size = int(input('Введите размер куска: '))

if ch_size % ch_len == 0 or ch_size % ch_wid == 0:
    print("Разлом можно совершить")
else:
    print("Разлом нельзя совершить")
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
inp_num = input("Введите целое положительное число: ")
if inp_num.isdigit():
    num = int(inp_num)
    if num > 0 or num <= 3999:
        result = ''
        values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
        numerals = ['M', 'CM', 'D', 'CD', 'C', 'XC', 'L', 'XL', 'X', 'IX', 'V', 'IV', 'I']
        for i in range(len(values)):
            while num >= values[i]:
                    result += numerals[i]
                    num -= values[i]
        print(f"Римское число: {result}")
    else:
        print("Число должно быть от 1 до 3999!")
else:
    print("Введены не числовые данные!")
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
def is_real_digit(s, negative=False):
    if not s:  
        return False
    
    has_decimal = False
    has_digits = False
    
    for i, char in enumerate(s):
        if char == '.':
            if has_decimal:
                return False
            has_decimal = True
        elif char in ['-', '+']:
            if i != 0:
                return False
        elif not ('0' <= char <= '9'):
            return False
        else:
            has_digits = True
    
    if not has_digits:
        return False

    if s[-1] == '.':
        return False
    
    if not negative and s and s[0] == '-':
        return False
    
    return True

inp_num = input("Введите число: ")
if is_real_digit(inp_num, negative=False):
    print("Является положительным вещественным числом")
else:
    if is_real_digit(inp_num, negative=True):
        print("Является отрицательным вещественным числом")
    else:
        print("Не является вещественным числом")
```
