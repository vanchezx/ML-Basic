## Задание 1: Получение однозначного числа

Пользователь вводит целое число, программа складывает все цифры числа, с полученным числом — то же самое, и так до тех пор, пока не получится однозначное число.

<br>

**Примеры:**
* `545   -> 5`
* `12345 -> 6`

````python
def sum_digit(num):
    num = abs(int(num))
    while num >= 10:
        total = 0
        while num > 0:
            total += num % 10
            num //= 10
        num = total
    return num

inp_num = input("Введите число: ")
if inp_num.isdigit():
    result = sum_digit(inp_num)
    print(f"Результат сложения числа {inp_num} = {result}")
else:
    print("Введены не числовые данные!")
````

## Задание 2: Кинотеатр

Дан список списков, каждый вложенный список состоит из 1 и 0, количество вложенных списков — количество рядов. Пользователь вводит, сколько билетов ему требуется. Программа должна найти ряд, где можно приобрести нужное количество билетов (места должны быть рядом). Если таких рядов несколько, то ближайший к экрану (ближайшим считается нулевой ряд). Если таких мест нет, то вывести False.

<br>

**Примеры:**
* `[[0,1,1,0], [1,0,0,0], [0,1,0,0]], 2 -> 1`
* `[[0,1,1,0], [1,0,1,0], [1,1,0,1]], 2 -> False`

````python
def find_seats(rows, needed_seats):
    for row_num, row in enumerate(rows):
        consecutive_free = 0
        for seat in row:
            if seat == 0:  # Свободное место
                consecutive_free += 1
                if consecutive_free >= needed_seats:
                    return row_num
            else:  # Занятое место
                consecutive_free = 0
    return False


#rows = [[0,1,1,0], [1,0,0,0], [0,1,0,0]]
rows = [[0,1,1,0], [1,0,1,0], [1,1,0,1]]
inp_tickets = input("Введите кол-во билетов: ")
if inp_tickets.isdigit():
    count = int(inp_tickets)
    result = find_seats(rows, count)
    if result:
        print(f"{count} билетов доступно на {result} ряду") 
    else:
        print("Билетов нет!")
else:
    print("Введены не верные данные!")
````

## Задание 3: Алгоритм RLE

Необходимо написать упрощенную версию алгоритма RLE. 
Алгоритм RLE объединяет подряд идущие символы в коэффициент и символ.

<br>

**Примеры:**
* `asssdddsssddd -> 1a3s3d3s3d`
* `aaabbbbccccc  -> 3a4b5c`
* `abcba         -> 1a1b1c1b1a`

````python
def rle(text):
    if not text:
        return ""
    
    encoded = []
    count = 1
    current_char = text[0]
    
    for i in range(1, len(text)):
        if text[i] == current_char:
            count += 1
        else:
            encoded.append(f"{count}{current_char}")
            current_char = text[i]
            count = 1
    encoded.append(f"{count}{current_char}")
    return "".join(encoded)


inp_text = input("Введите данные: ")
result = rle(inp_text)
print(f"'{inp_text}' -> '{result}'")
````

## Задание 4: Шифр Цезаря

Реализуйте программу для шифрования текста с помощью шифра Цезаря.  
Шифр Цезаря — это метод шифрования, при котором каждая буква в тексте заменяется буквой, стоящей на фиксированное число позиций дальше в алфавите.

### Требования
1. Программа должна принимать на вход:
   - Строку текста для шифрования
   - Ключ шифра (число от 0 до 25)

2. Правила шифрования:
   - Шифруются только латинские буквы (a-z, A-Z)
   - Сохраняется регистр букв (заглавные остаются заглавными)
   - Пробелы, цифры и знаки препинания остаются без изменений
   - Сдвиг циклический (после 'z' идет 'a', после 'Z' идет 'A')

### Примеры

* `"Hello World", 3      -> "Khoor Zruog"`
* `"Python is great!", 5 -> "Udymts nx lwjfy!"`
* `"XYZ abc", 3          -> "ABC def"`
* `"Test 123", 1         -> "Uftu 123"`

### Алгоритм
* Для каждого символа в строке:
   - Если это латинская буква, найти её позицию в алфавите
   - Добавить к позиции значение ключа
   - Если результат больше 25, использовать остаток от деления на 26
   - Преобразовать обратно в букву, сохранив регистр
   - Если это не буква, оставить символ без изменений


````python
def csr_encrypt(text, shift):
    result = []
    shift = shift % 26
    
    for char in text:
        if 'a' <= char <= 'z':
            result.append(chr((ord(char) - ord('a') + shift) % 26 + ord('a')))
        elif 'A' <= char <= 'Z':
            result.append(chr((ord(char) - ord('A') + shift) % 26 + ord('A')))
        else:
            result.append(char)
    return ''.join(result)

inp_text = input("Введите данные для шифрования: ")
inp_keys = input("Введите ключ шифрования: ")
if inp_keys.isdigit():
    crypted_text = csr_encrypt(inp_text, int(inp_keys))
    print(f"Зашифрованные данные:  {crypted_text}")   
else:
    print("Введены не верные данные ключа шифрования!")
````

## Задание 5: Табель успеваемости

Пользователь в бесконечном цикле (пока не будет введена пустая строка) вводит строки вида: 'название предмета' 'фамилия ученика' 'оценка'. После окончания ввода программа выводит в консоль название предмета, далее список учеников и все их оценки в виде таблицы.

<br>

**Примеры:**

<br>

*Ввод:*
```python
Математика Иванов  5
Математика Иванов  4
Литература Иванов  3
Математика Петров  5
Литература Сидоров 3
Литература Петров  5
Литература Иванов  4
Математика Сидоров 3
Математика Петров  5
```

<br>

*Результат:*
```python
Математика # вводим название предмета

# выводим список учеников и их оценки

Сидоров 3
```

```python
Литература # вводим название предмета

# выводим список учеников и их оценки
Иванов  3 4
Сидоров 3
Петров  5
```

````python
print('=' * 59)
print('Введите данные в формате: "ПРЕДМЕТ ФАМИЛИЯ ОЦЕНКА"')
print('Для завершения ввода данных введите пустую строку....')
print('=' * 59)

subjects = {}

while True:
    line = input().strip()
            
    if not line:
        break
                
    parts = line.split()
    if len(parts) != 3:
        print('Ошибка! Введены некорректные данные')
        print('Введите данные в формате: "ПРЕДМЕТ ФАМИЛИЯ ОЦЕНКА", оценка от 1 до 5')
        continue
                
    subject = parts[0]
    student = parts[1]
    grade = parts[2]
            
    if not grade.isdigit() or not (1 <= int(grade) <= 5):
        print(f"Ошибка: Введена некорректная оценка '{grade}'")
        print('Введите данные в формате: "ПРЕДМЕТ ФАМИЛИЯ ОЦЕНКА", оценка от 1 до 5')
        continue
            
    if subject not in subjects:
        subjects[subject] = {}
            
    if student not in subjects[subject]:
        subjects[subject][student] = []
            
    subjects[subject][student].append(grade)


print('=' * 59)
print(' ' * 20+ 'ТАБЕЛЬ УСПЕВАЕМОСТИ' +' ' * 20)
print('=' * 59)

for subject in sorted(subjects.keys()):
    print(f"\n{subject}")
    print("-" * len(subject))

    students_data = subjects[subject]
        
    for student in sorted(students_data.keys()):
        grades = students_data[student]
        grades_str = " ".join(grades)
        print(f"{student:<15} {grades_str}")
        
    print()
````
