## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
def choose_case(text, text_format=None):
    if text_format == 'snake':
        return to_snake_case(text)
    elif text_format == 'pascal':
        return to_pascal_case(text)
    else:
        if '_' in text:
            return to_pascal_case(text)
        else:
            return to_snake_case(text)

def to_pascal_case(snake_text):
    if not snake_text:
        return ""
    words = snake_text.split('_')
    processed_words = []
    for word in words:
        if word:
            processed_words.append(word.capitalize())
    return ''.join(processed_words)

def to_snake_case(pascal_text):
    if not pascal_text:
        return ""
    result = []
    for i, char in enumerate(pascal_text):
        if char.isupper() and i > 0:
            result.append('_')
        result.append(char.lower())
    return ''.join(result)

inp_text = input("Введите данные: ")
inp_form = input("Введите формат данных или оставьте поле пустым: ")
result = choose_case(inp_text,inp_form)
print(f"{inp_text} -> {result}")
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
def is_valid_date(date_str):
    parts = date_str.split('.')
    if len(parts) != 3:
        return False
            
    day, month, year = parts
        
    if not (day.isdigit() and month.isdigit() and year.isdigit()):
        return False
        
    day = int(day)
    month = int(month)
    year = int(year)
        
    if year < 1 or year > 9999:
        return False
        
    if month < 1 or month > 12:
        return False
        
    days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        
    if month == 2 and is_leap_year(year):
        days_in_feb = 29
    else:
        days_in_feb = days_in_month[month - 1]
        
    if month == 2:
        max_days = days_in_feb
    else:
        max_days = days_in_month[month - 1]
        
    if day < 1 or day > max_days:
        return False
        
    return True

def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

inp_date = input("Введите дату: ")
if is_valid_date(inp_date):
    print("Дата валидна!")
else:
    print("Дата не валидна!")
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def is_simple(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, n, 2):
        if n % i == 0:
            return False
    return True

inp_num = input("Введите число: ")
if inp_num.isdigit():
    if is_simple(int(inp_num)):
        print("Является простым числом")
    else:
        print("Не является простым числом")
else:
    print("Введены не числовые данные!")
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
def validate_name(name):
    if not name or not name.replace(' ', '').isalpha():
        return None
    formatted_name = ' '.join(word.capitalize() for word in name.split())
    return formatted_name

def validate_age(age_str):
    age = int(age_str)
    if 18 <= age <= 60:
        return age
    else:
        return None
        
def validate_id(user_id, existing_ids):
    user_id_int = int(user_id)
    formatted_id = str(user_id_int).zfill(8)
    if formatted_id in existing_ids:
        print("Ошибка: ID должен быть уникальным!")
        return None 
    return formatted_id

users = {}  
print('Введите данные ползователя....')
print('Для завершения ввода данных введите пустую строку....')
print("-" * 60)
    
while True:
    print(f"\nПользователь #{len(users) + 1}")
    
    inp_name = input("Введите имя: ").strip()
    if not inp_name:
        break
            
    inp_surname = input("Введите фамилию: ").strip()
    if not inp_surname:
        break
            
    inp_age = input("Введите возраст: ").strip()
    if not inp_age:
        break
            
    inp_id = input("Введите ID: ").strip()
    if not inp_id:
        break
        
    form_name = validate_name(inp_name)
    if not form_name:
        print("Ошибка: имя должно содержать только буквы!")
        continue
            
    form_surname = validate_name(inp_surname)
    if not form_surname:
        print("Ошибка: фамилия должна содержать только буквы!")
        continue
            
    form_age = validate_age(inp_age)
    if not form_age:
        print("Ошибка: возраст должен быть числом от 18 до 60!")
        continue
            
    form_id = validate_id(inp_id, users.keys())
    if not form_id:
        print("Ошибка: ID должен быть целым числом и уникальным!")
        continue
        
    users[form_id] = (form_name, form_surname, form_age)
    print(f"Пользователь {form_name} {form_surname} успешно добавлен!")

if users:
    print("\n" + "="*60)
    print(f"{'ID':<10} {'Имя':<15} {'Фамилия':<15} {'Возраст':<10}")
    print("-" * 60)
    
    for user_id, user_data in users.items():
        name, surname, age = user_data
        print(f"{user_id:<10} {name:<15} {surname:<15} {age:<10}")
    
    print("=" * 60)
    print(f"Всего пользователей: {len(users)}")
else:
    print("Нет данных для отображения!")
```
