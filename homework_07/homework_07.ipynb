# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr = np.zeros(10)
arr[4] = 1
arr_2d = arr.reshape(2, 5)
print("Одномерный массив:")
print(arr)
print("Двумерный массив:")
print(arr_2d)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr = np.arange(10, 50)
arr_reversed = arr[::-1]
elements = arr_reversed[arr_reversed % 2 == 0]

print("Исходный массив:")
print(arr)
print("Развернутый массив:")
print(arr_reversed)
print("Четные элементы:")
print(elements)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr = np.arange(9).reshape(3, 3)
print(arr)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr = np.random.rand(4, 3, 2)
min_value = np.min(arr)
max_value = np.max(arr)

print("Массив 4x3x2:")
print(arr)
print(f"Минимальное значение: {min_value}")
print(f"Максимальное значение: {max_value}")
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr1 = np.random.randint(0, 10, (6, 4))
arr2 = np.random.randint(0, 10, (4, 3))
result = np.dot(arr1, arr2)  # или arr1 @ arr2

print("Первый массив (6x4):")
print(arr1)
print("Второй массив (4x3):")
print(arr2)
print("Результат матричного умножения:")
print(result)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr = np.random.randint(0, 50, (7, 7))
mean = np.mean(arr)
std = np.std(arr)
arr_normalized = (arr - mean) / std

print("Исходный массив:")
print(arr)
print(f"Среднее значение: {mean}")
print(f"Стандартное отклонение: {std}")
print(f"Нормализованный массив:")
print(arr_normalized)
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
median_bill = tips['total_bill'].median()
print(f"Медиана: {median_bill}")
filtered_tips = tips[tips['total_bill'] > median_bill]
avg_tip_by_sex = filtered_tips.groupby('sex')['tip'].mean()
print("Средние чаевые по полу:")
print(avg_tip_by_sex)
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
sns.histplot(data=tips, x='total_bill', bins=30, kde=True)
plt.title('Распределение суммы счета')
plt.xlabel('Сумма счета')
plt.ylabel('Частота')
plt.grid(True, alpha=0.5)
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title('Взаимосвязь между суммой счета и чаевыми')
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(tips)
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Распределение суммы счета по дням недели')
plt.xlabel('День недели')
plt.ylabel('Сумма счета')
plt.grid(True, alpha=0.3)
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
sns.histplot(data=tips, x='tip', hue='time', multiple='stack', bins=20)
plt.title('Распределение чаевых от времени суток')
plt.xlabel('Чаевые')
plt.ylabel('Частота')
plt.grid(True, alpha=0.3)
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
# Один график с разделением по столбцам (col)
g = sns.FacetGrid(tips, col='sex', height=6)
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip', hue='smoker', palette={'Yes': 'red', 'No': 'blue'})
g.add_legend(title='Курильщики')
g.set_axis_labels('Сумма счёта', 'Чаевые')
for ax, title in zip(g.axes.flat, ['Мужчины', 'Женщины']):
    ax.set_title(title)

plt.show()
```
